```python
from crewai import LLM

llm = LLM(model="gemini/gemini-2.0-flash", temperature=0)
response = llm.call("Write a Python function that returns the Fibonacci sequence up to n.")
```

```python
from crewai import Agent, Task, Crew


email_assistant = Agent(
    role="Email Assistant Agent",
    goal="Improve emails and make them sound professional",
    backstory="A highly experienced communication expert skilled in crafting clear and effective messages.",
    verbose=True,
    llm=llm
)

original_email = """
Hey team, just wanted to check in and see how everyone's doing with the project. Let me know if you need any help or have any questions. Lets get drunk!
"""

email_task = Task(
    description="Improve the following email to make it sound more professional:\n" + original_email,
    agent=email_assistant,
    expected_output="A professional written email with proper formatting"
)

crew = Crew(
    agents=[email_assistant],
    tasks=[email_task],
    verbose=True
)

result = crew.kickoff()
print(result)
```
